```python
# Import packages
from pyspark.sql import SparkSession
import time

# Create a SparkSession
spark = SparkSession.builder.appName("SparkSQL").getOrCreate()
```

```python
# 1. Read the home_sales_revised.csv from the provided AWS S3 bucket location into a PySpark DataFrame.
from pyspark import SparkFiles

# Define the URL where the CSV file is located
url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.2/22-big-data/home_sales_revised.csv"

# Add the file to Spark's context
spark.sparkContext.addFile(url)
```

```python
# 2. Create a temporary view of the DataFrame.
df = spark.read.option("header","true").option("inferSchema","true").csv("file://" + SparkFiles.get("home_sales_revised.csv"))
# Show the first 5 rows of the DataFrame
df.show(5)
```

```python
# Create a temporary view of the DataFrame called 'home_sales'
df.createOrReplaceTempView("home_sales")
```

```python
# 3. What is the average price for a four bedroom house sold per year, rounded to two decimal places?
query = """
SELECT
    YEAR(date) AS year,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
WHERE bedrooms = 4
GROUP BY YEAR(date)
ORDER BY year;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()
```

```python
# 4. What is the average price of a home for each year the home was built,
# that have 3 bedrooms and 3 bathrooms, rounded to two decimal places?
query = """
SELECT
    date_built AS year_built,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
WHERE bedrooms = 3 AND bathrooms = 3
GROUP BY date_built
ORDER BY year_built;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()
```

```python
# 5. What is the average price of a home for each year the home was built,
# that have 3 bedrooms, 3 bathrooms, with two floors,
# and are greater than or equal to 2,000 square feet, rounded to two decimal places?
query = """
SELECT
    date_built AS year_built,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
WHERE bedrooms = 3 AND bathrooms = 3 AND floors = 2 AND sqft_living >= 2000
GROUP BY date_built
ORDER BY year_built;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()
```

```python
# 6. What is the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000? Order by descending view rating.
# Although this is a small dataset, determine the run time for this query.

start_time = time.time()
query = """
SELECT
    view,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
GROUP BY view
HAVING average_price >= 350000
ORDER BY view DESC;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 7. Cache the temporary table home_sales.
spark.sql("CACHE TABLE home_sales")
```

```python
# 8. Check if the table is cached.
spark.catalog.isCached('home_sales')
```

```python
# Examining the Cached Data
spark.sql("SELECT * FROM home_sales LIMIT 5").show()
```

```python
# 9. Using the cached data, run the last query above, that calculates
# the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000.
# Determine the runtime and compare it to uncached runtime.

start_time = time.time()
query = """
SELECT
    view,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
GROUP BY view
HAVING average_price >= 350000
ORDER BY view DESC;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 10. Partition by the "date_built" field on the formatted parquet home sales data
# Define the output path (local or cloud storage)
output_path = "/content/drive/My Drive/home_sales_partitioned"

# Save DataFrame as Parquet format, partitioned by date_built
df.write.mode("overwrite").partitionBy("date_built").parquet(output_path)

print("Parquet file successfully saved to Google Drive with partitioning by date_built.")
```

```python
# 11. Read the parquet formatted data.
partitioned_df.filter(partitioned_df.date_built == 2015).show(5)
```

```python
# 12. Create a temporary table for the parquet data.
partitioned_df.createOrReplaceTempView("home_sales_parquet")

print("Temporary table 'home_sales_parquet' created successfully.")
```

```python
# 13. Using the parquet DataFrame, run the last query above, that calculates
# the average price of a home per "view" rating, rounded to two decimal places,
# having an average home price greater than or equal to $350,000.
# Determine the runtime and compare it to the cached runtime.

start_time = time.time()

query = """
SELECT
    view,
    ROUND(AVG(price), 2) AS average_price
FROM home_sales
GROUP BY view
HAVING average_price >= 350000
ORDER BY view DESC;
"""
# Execute the query
result = spark.sql(query)
# Display the result
result.show()

print("--- %s seconds ---" % (time.time() - start_time))
```
